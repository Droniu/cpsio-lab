### Configuration

```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np

# plot size
plt.rcParams["figure.figsize"] = (8.5,4)
```

### Single plot
The cell bellow illustrates data from `ekg100.txt` file.

```python
# data loading
data = np.fromfile('data/ekg100.txt', sep=' ')

# given that Fs = 360 Hz
SAMPLING_RATE = 360

X_range = np.arange(stop=(len(data)))
X = X_range / SAMPLING_RATE

# plt.plot(X, data)
# plt.xlabel("Time [s]")
# plt.ylabel("Voltage [mV]")
# plt.show()


fig1 = plt.figure(1)

ax1 = fig1.add_subplot(111)
ax1.set_xlabel("Time [s]")
ax1.set_ylabel("Voltage [mV]")
ax1.plot(X, data)

fig1.canvas.draw()
```

## Sine signal

```python

plt.clf()
FS = 50 #Hz
LENGTH = 65536
SAMPLES = np.arange(0, LENGTH) / FS


SINE = np.sin(SAMPLES)

#plt.plot(SAMPLES, SINE)
#plt.show()

fig2 = plt.figure(2)
ax2 = fig2.add_subplot(111)
ax2.plot(SAMPLES, SINE)

fig2.canvas.draw()
```

## Fourier transform of signal

```python
FFT = np.fft.fft(SINE)

X_FFT = np.linspace(0, FS/2)

plt.plot(np.abs(FFT))
plt.show()
```
